```python
# %load Chapter10/listing10_6.py
# numberlines.py                                   #  1 #  1
                                                   #  2 #  2
import fileinput                                   #  3 #  3
                                                   #  4 #  4
for line in fileinput.input(inplace=True):         #  5 #  5
    line = line.rstrip()                           #  6 #  6
    num  = fileinput.lineno()                      #  7 #  7
    print('{:<50} # {:2d}'.format(line, num))      #  8 #  8
```

```python
#静态变量需要采用类名.变量名的形式来处理

class myaddcounter:
    count=0
    def __init__(self):
        myaddcounter.count+=1
        
    
```

```python
#通过固定给某个文件记录数据，来统计执行次数。
ct=myaddcounter()
f=open('Chapter11/listing11_2.txt','a')
info='\nrjb add info now.'
info=info + str(ct.count)
f.write(info)
f.close()
```
